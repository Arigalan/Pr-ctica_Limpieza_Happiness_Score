Paso 6: Rellenar con media

```python
df["happiness_score"] = df["happiness_score"].fillna(df["happiness_score"].mean())
df
```

¿Este método representa bien los datos?

R: Podemos ver que si es la más adecuada, debido a que los datos van escalando, por lo cual no hay valores outliers o muy alejados de la media, por lo cual ¿, no convendría usar la mediana. Ahora, como hay valores repetidos casi la misma cantidad de veces, pues no conviene usar la moda porque eso no representa a los datos

Paso 7: Rellenar con mediana

```python
df["happiness_score"] = df["happiness_score"].fillna(df["happiness_score"].median())
df
```

¿Cuál método parece más adecuado?

R: Me parece más adecuado el de la media

Paso 8: Comparación

```python
print("La media es: ", df['happiness_score'].mean())
print("La mediana es: ", df['happiness_score'].median())
```

¿Hay diferencias entre media y mediana?

R: No hay mucha diferencia
